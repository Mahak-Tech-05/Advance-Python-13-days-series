```python
import sqlite3

# Creating connection (this creates file DB in Colab)
con = sqlite3.connect("student.db")
cursor = con.cursor()

print("Database Connected Successfully!")

# Creating table
cursor.execute("""
CREATE TABLE IF NOT EXISTS student_details(
    ID INTEGER PRIMARY KEY AUTOINCREMENT,
    Name TEXT,
    Age INTEGER,
    Branch TEXT,
    Year INTEGER,
    Semester INTEGER
)
""")
con.commit()

print("Table created successfully!")

# Add Student
def add_student():
    name = input("Enter your name: ")
    age = int(input("Enter your age: "))
    branch = input("Enter your branch: ")
    year = int(input("Enter your year: "))
    semester = int(input("Enter your semester: "))

    query = "INSERT INTO student_details(Name, Age, Branch, Year, Semester) VALUES (?, ?, ?, ?, ?)"
    values = (name, age, branch, year, semester)

    cursor.execute(query, values)
    con.commit()

    print("Student added successfully!")

# View Students
def view_student():
    cursor.execute("SELECT * FROM student_details")
    records = cursor.fetchall()

    print("----- Student Records -----")
    for row in records:
        print(row)

# Delete Student
def delete_student():
    id = int(input("Enter ID to delete: "))
    cursor.execute("DELETE FROM student_details WHERE ID = ?", (id,))
    con.commit()
    print("Deleted successfully!")

# Update Student
def update_student():
    id = int(input("Enter ID to update: "))

    print("1.Name\n2.Age\n3.Branch\n4.Year\n5.Semester")
    choice = input("Enter choice: ")

    if choice == '1':
        new = input("Enter new name: ")
        cursor.execute("UPDATE student_details SET Name=? WHERE ID=?", (new, id))

    elif choice == '2':
        new = int(input("Enter new age: "))
        cursor.execute("UPDATE student_details SET Age=? WHERE ID=?", (new, id))

    elif choice == '3':
        new = input("Enter new branch: ")
        cursor.execute("UPDATE student_details SET Branch=? WHERE ID=?", (new, id))

    elif choice == '4':
        new = int(input("Enter new year: "))
        cursor.execute("UPDATE student_details SET Year=? WHERE ID=?", (new, id))

    elif choice == '5':
        new = int(input("Enter new semester: "))
        cursor.execute("UPDATE student_details SET Semester=? WHERE ID=?", (new, id))

    else:
        print("Invalid choice")
        return

    con.commit()
    print("Updated successfully!")

# Main Menu
while True:
    print("\n------ MAIN MENU ------")
    print("1. Add Student")
    print("2. Update Student")
    print("3. View Students")
    print("4. Delete Student")
    print("5. Exit")

    c = input("Enter your choice: ")

    if c == '1':
        add_student()

    elif c == '2':
        update_student()

    elif c == '3':
        view_student()

    elif c == '4':
        delete_student()

    elif c == '5':
        print("Exiting...")
        break

    else:
        print("Invalid choice")
```
