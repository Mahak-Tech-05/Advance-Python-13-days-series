```python
import sqlite3
import ipywidgets as widgets
from IPython.display import display, clear_output

# ------- Database connection ------
conn = sqlite3.connect("student.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS student(
    ID INTEGER PRIMARY KEY AUTOINCREMENT,
    Name TEXT,
    age INTEGER,
    course TEXT
)
""")
conn.commit()

# -------- Widgets (GUI) ----------
name_input = widgets.Text(description="Name:")
age_input = widgets.Text(description="Age:")
course_input = widgets.Text(description="Course:")

submit_btn = widgets.Button(description="Add Data", button_style='success')
output = widgets.Output()

# -------- Insert Function ----------
def insert_data(b):
    with output:
        clear_output()

        name = name_input.value
        age = age_input.value
        course = course_input.value

        if name == "" or age == "" or course == "":
            print("❌ All fields are required")
            return

        cursor.execute(
            "INSERT INTO student (Name, age, course) VALUES (?, ?, ?)",
            (name, age, course)
        )
        conn.commit()

        print("✅ Data Inserted Successfully")

        # Clear fields
        name_input.value = ""
        age_input.value = ""
        course_input.value = ""

submit_btn.on_click(insert_data)

# Display GUI
display(name_input, age_input, course_input, submit_btn, output)
```
