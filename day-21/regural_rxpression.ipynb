```python
# Importing regex module
import re
```

```python
text = "Hello my email is test123@gmail.com and my number is 9876543210"
```

```python
print("----- re.match() -----")

result = re.match("Hello", text)

if result:
    print("Match found:", result.group())
else:
    print("No match found")
```

```python
print("\n----- re.search() -----")

result = re.search("email", text)

if result:
    print("Found:", result.group())
```

```python
print("\n----- re.findall() -----")

numbers = re.findall(r"\d+", text)

print("Numbers found:", numbers)
```

```python
print("\n----- re.sub() -----")

new_text = re.sub("Hello", "Hi", text)

print("Updated text:", new_text)
```

```python
print("\n----- Character Set Example -----")

result = re.findall(r"[A-Za-z]+", text)

print("Words only:", result)
```

```python
print("\n----- Email Validation -----")

email = "student123@gmail.com"

pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"

if re.match(pattern, email):
    print("Valid Email")
else:
    print("Invalid Email")
```
