```python
# ==============================
# 📊 STUDENT DATA ANALYSIS PROJECT
# ==============================

import pandas as pd

# ---- Step 1: Create Data ----
data = {
    "Name": ["Ram", "Basanti", "Thakur", "Gita", "Shamelal"],
    "Marks": [85, 90, None, 40, 75],
    "Age": [20, 21, 19, 18, None]
}

df = pd.DataFrame(data)

print("----- Original Data -----")
display(df)   # better than print in Colab

# ---- Step 2: Data Cleaning ----
df["Marks"] = df["Marks"].fillna(df["Marks"].mean().round(2))
df["Age"] = df["Age"].fillna(df["Age"].mean().round(0))

print("\n----- After Cleaning -----")
display(df)

# ---- Step 3: Basic Analysis ----
print("\n----- Analysis -----")
print("Average Marks:", df["Marks"].mean().round(2))
print("Maximum Marks:", df["Marks"].max())
print("Minimum Marks:", df["Marks"].min())

# ---- Step 4: Result Column ----
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 50 else "Fail")

# ---- Step 5: Grade System (Advanced) ----
def grade(marks):
    if marks >= 80:
        return "A"
    elif marks >= 60:
        return "B"
    elif marks >= 50:
        return "C"
    else:
        return "F"

df["Grade"] = df["Marks"].apply(grade)

print("\n----- Final Data with Result & Grade -----")
display(df)

# ---- Step 6: Filter Passed Students ----
passed_students = df[df["Result"] == "Pass"]

print("\n----- Students Who Passed -----")
display(passed_students)

# ---- Step 7: Save to CSV ----
file_name = "student_output.csv"
df.to_csv(file_name, index=False)

print("\n✅ Data saved to", file_name)

# ---- Step 8: Download File (Colab only) ----
from google.colab import files
files.download(file_name)
```
